### Cleaning data

```python
#SB
# Change datatype
def change_datatype(data):
    data = data.astype({'Loan Date End': 'datetime64', 'Joined': 'datetime64'})
    return data

#data = change_datatype(data)
```

```python
# Convert String to float for inches of column Height
def parse_ht(ht):
    # format: 7' 0.0"
    ht_ = ht.split("'")
    ft_ = float(ht_[0])
    in_ = float(ht_[1].replace("\"",""))
    return (12*ft_) + in_

def clean_height(data):
    data2 = data.copy()
    data2["Height"] = data2["Height"].apply(lambda x:parse_ht(x))
    return data2

#data = clean_height(data)
```

```python
# Rename some values of column Nationality
def replace_nation_name(data):
    data2 = data.copy()
    data2["Nationality"].replace({"Trinidad &amp; Tobago": "Trinidad & Tobago", "Antigua &amp; Barbuda": "Antigua & Barbuda", "São Tomé &amp; Príncipe": "São Tomé & Príncipe"}, inplace=True)
    return data2

#data2 = replace_nation_name(data)
```

```python
# Group Nationality
def group_nationality(x):
    if x in ['Albania', 'Austria', 'Azerbaijan', 'Belarus', 'Belgium', 'Bosnia Herzegovina', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 'England', 'Estonia', 'Faroe Islands', 'Finland', 'France', 'Georgia', 'Germany', 'Gibraltar', 'Greece', 'Hungary', 'Iceland', 'Italy', 'Kosovo', 'Latvia', 'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 'Moldova', 'Montenegro', 'Netherlands', 'North Macedonia', 'Northern Ireland', 'Norway', 'Poland', 'Portugal', 'Republic of Ireland', 'Romania', 'Russia', 'Scotland', 'Serbia', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'Ukraine', 'Wales']:
        return 'Europe'
    elif x in ['Afghanistan', 'Armenia', 'China PR', 'Chinese Taipei', 'Cyprus', 'Georgia', 'Hong Kong', 'India', 'Indonesia', 'Iran', 'Iraq', 'Israel', 'Japan', 'Jordan', 'Kazakhstan', 'Korea DPR', 'Korea Republic', 'Lebanon', 'Macau', 'Malaysia', 'Palestine', 'Philippines', 'Qatar', 'Russia', 'Saudi Arabia', 'Singapore', 'Syria', 'Thailand', 'Turkey', 'United Arab Emirates', 'Uzbekistan', 'Vietnam']:
        return 'Asia'
    elif x in ['Algeria', 'Angola', 'Benin', 'Burkina Faso', 'Burundi', 'Cameroon', 'Cape Verde', 'Central African Republic', 'Chad', 'Comoros', 'Congo', 'DR Congo', 'Egypt', 'Equatorial Guinea', 'Eritrea', 'Gabon', 'Gambia', 'Ghana', 'Guinea', 'Guinea Bissau', 'Ivory Coast', 'Kenya', 'Liberia', 'Libya', 'Madagascar', 'Malawi', 'Mali', 'Mauritania', 'Morocco', 'Mozambique', 'Namibia', 'Niger', 'Nigeria', 'Rwanda', 'São Tomé & Príncipe', 'Senegal', 'Sierra Leone', 'South Africa', 'South Sudan', 'Sudan', 'Tanzania', 'Togo', 'Tunisia', 'Uganda', 'Zambia', 'Zimbabwe']:
        return 'Africa'
    elif x in ['Antigua & Barbuda', 'Barbados', 'Belize', 'Canada', 'Costa Rica', 'Cuba', 'Dominican Republic', 'El Salvador', 'Grenada', 'Haiti', 'Honduras', 'Jamaica', 'Mexico', 'Panama', 'Puerto Rico', 'Saint Kitts and Nevis', 'Trinidad & Tobago', 'United States']:
        return 'North America'
    elif x in ['Aruba', 'Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Curacao', 'Ecuador', 'Guyana', 'Paraguay', 'Peru', 'Uruguay', 'Venezuela']:
        return 'South America'
    elif x in ['Australia', 'Guam', 'New Zealand', 'New Caledonia', 'Papua New Guinea']:
        return 'Australia and Oceania'
    else:
        print(x)
        return x

def clean_nationality(data):
    data2 = data.copy()
    data2['Nationality'] = list(map(group_nationality, data2['Nationality']))
    return data

#data2 = clean_nationality(data)
```

```python
# Convert categorical columns to numerical
def convert_to_num(data):
    data2 = data.copy()
    cols_to_clean = ['W/F','SM','IR','A/W','D/W']    
    for col in cols_to_clean:
        if (col in ['W/F','SM','IR']):        
            #Remove rating star and change dtype to int
            data2[col] = data2[col].apply(lambda x: x.replace('★','')).astype(int)
        else:
            #Changing categorical values in numerical values
            ord_dict = {'Low':0,'Medium':1,'High':2}
            data2[col] = data2[col].map(ord_dict)
    return data2

#data3 = convert_to_num(data)
#data3[['W/F','SM','IR','A/W','D/W']].head()
```

```python
# Drop columns
def drop_columns(data):
    data2 = data.copy()
    data2 = data2.drop(['Loan Date End', 'Name', 'Club', 'Position', 'Joined', 'Team & Contract', 'foot'], axis=1)
    return data2

#data4 = drop_columns(data)
#data4.head()
```

```python
# Clean Wage
def Wage(amount): 
    if (amount[-1]=="M"):
        return float(amount[1:-1])*1000000 
    elif(amount[-1]=="K"):
        return float(amount[1:-1])*1000
    else:
        return float(amount[1:])

#print(Wage("€1M"))

def clean_wage(data):
   # data = data.copy()
    cols_to_clean = ['Wage','Value','Release Clause']
    for col in cols_to_clean:
        data[col] = data[col].apply(lambda x: Wage(x))
        #data[col]=map(Wage, data[col])
    return data

# print(data)
def clean_dataset(data):
    data2 = data.copy()
    data2 = clean_wage(data)
   # df = clean_weight(df)
    return data2

#print(df)
#data = clean_dataset(data)
#data
```

```python
def final_cleaning(data):
    data = change_datatype(data)
    data = clean_height(data)
    data = replace_nation_name(data)
    data = clean_nationality(data)
    data = convert_to_num(data)
    data = drop_columns(data)
    data = clean_dataset(data)
    return data

data_clean = final_cleaning(data)
data_clean.head()
```
