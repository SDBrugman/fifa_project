```python
# Convert String to float for inches of column Height
def parse_ht(ht):
    # format: 7' 0.0"
    ht_ = ht.split("'")
    ft_ = float(ht_[0])
    in_ = float(ht_[1].replace("\"",""))
    return (12*ft_) + in_

data["Height"].apply(lambda x:parse_ht(x))
```
