```python
# Convert categorical columns to numerical
def convert_to_num(data):
    data2 = data.copy()
    cols_to_clean = ['W/F','SM','IR','A/W','D/W']    
    for col in cols_to_clean:
        if (col in ['W/F','SM','IR']):        
            #Remove rating star and change dtype to int
            data2[col] = data2[col].apply(lambda x: x.replace('★','')).astype(int)
        else:
            #Changing categorical values in numerical values
            ord_dict = {'Low':0,'Medium':1,'High':2}
            data2[col] = data2[col].map(ord_dict)
    return data2

data3 = convert_to_num(data)
data3[['W/F','SM','IR','A/W','D/W']].head()
```

```python
# Drop column loan Date End - too many NaN values and not necessary for predictions
# Drop column Club and Position also
def drop_columns(data):
    data2 = data.copy()
    data2 = data2.drop(['Loan Date End', 'Name', 'Club', 'Position', 'Joined', 'Team & Contract', 'foot'], axis=1)
    return data2

data4 = drop_columns(data)
data4.head()
```
